```python
stops2 = ['LGBTQ', 'mental', 'health']
text2 = df2['Headline'].to_string(index=False)
for item in stops2:
    text2 = text2.replace(item, " ")
#text = text.replace(stops, " ")

print("Remove keywords:")

#print(type(text2))
text_list2 = text2
#print(type(text_str))

text_list2=text_list2.split()

print("Change to list:")
#print(type(text_list))

print("Length of text_list original:", len(text_list2))


count = 0
for word in text_list2:
    text_list2[count] = text_list2[count].replace('...', '')
    count += 1

for word in text_list2:
    if len(word) <=2:
        #print(word)
        text_list2.remove(word)

stops2 = stopwords.iloc[:,0].tolist()
print("Length of stops:", len(stops2))

nostops2 = [i for i in text_list2 if i not in stops2]
print("compare lengths before and after removing stopwords")
print(len(text_list2))
print(len(nostops2))


nostops2 = ' '.join(nostops2)
```
