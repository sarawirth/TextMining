```python
import pandas as pd
import requests 
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
#Website 1: Springer article: anti-gender affirming care

url = 'https://link.springer.com/article/10.1007/s11930-023-00358-x#:~:text=Three%20recent%20papers%20examined%20the,today\'s%20adolescents%20is%20net%20beneficial.'

r = requests.get(url)
  
# check status code for response received 
# success code - 200 
print(r) 
  
# print content of request 

#print(r.text)
```

```python

tree = BeautifulSoup(r.text)
good_html = tree.prettify()
#tree
#good_html
```

```python
html = urlopen(url).read()
soup = BeautifulSoup(html, features="html.parser")

# kill all script and style elements
for script in soup(["script", "style"]):
    script.extract()    # rip it out

#soup
# get text
text = soup.get_text()
#print(text)
```

```python
#cleaning text
# break into lines and remove leading and trailing space on each
lines = (line.strip() for line in text.splitlines())
#print(text)
# break multi-headlines into a line each
chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
```

```python
# drop blank lines
text = '\n\n'.join(chunk for chunk in chunks if chunk)
#textdf = pd.DataFrame(text)
#print(textdf)
#print(type(text))
print(text)

```

```python
file1 = open('web1text.txt', 'w')
file1.close()

with open('web1text.txt', 'w') as f:
    f.writelines(text)

# Using readlines()
file1 = open('web1text.txt', 'r')
Lines = file1.readlines()
#print(len(Lines))

count = 0
# Strips the newline character
for line in Lines:
    print("Line{}: {}".format(count, line.strip()))
    count += 1
    #print(line.strip())
```
